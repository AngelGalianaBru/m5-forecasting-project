Verificar instalación de Java y PySpark

```python
import findspark
# findspark localiza la instalación de pyspark que descargó pip 
# y añade los binarios de Spark al sistema en tiempo de ejecución.
findspark.init()

from pyspark.sql import SparkSession

# Levantamos la sesión
spark = SparkSession.builder \
    .appName("M5_Setup_Test") \
    .master("local[*]") \
    .getOrCreate()

print(f"Versión de Spark corriendo: {spark.version}")

# Test rápido de procesamiento
df_test = spark.createDataFrame([("Hola", 1), ("Mundo", 2)], ["Palabra", "ID"])
df_test.show()
```
